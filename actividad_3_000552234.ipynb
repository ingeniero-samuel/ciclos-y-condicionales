<h3>Analisis de filtracion de señales:</h3>

formula: funcion de densidad de probabilidad de la campana de gauss<br>
o=300Hz<br>
u=1000Hz<br>
PI=3.1416<br>

fdp = (1/(o*sqrt(2*PI))*e^(-1/2)*((x-u)/o)^2)<br>

    program filtracion_de_señales

    Declare float fdp, porcent,
    constant U=1000, O=300, PI = 3.1416, e = 2.71828

    Input n
    
    porcent = 0

````python
import math
n=int(input("ingrese la frecuencia de la señal hasta donde usted desee: "))
o, u , PI, e= 300, 1000, 3.1416, 2.71828

porcent = 0
x=0

for i in range(n):
  x+=1
  fdp = (1 / (o * math.sqrt(2 * PI))) * e ** (-1/2 * ((x - u) / o) ** 2)
  porcent += fdp
  porcent_total = porcent
print(f"la densidad de probabilidad de filtracion de señal es: {porcent_total}")

if porcent_total >0.30:
  print("por ende la señal si pasa")
else:
  print("por ende esta se encuentra bloqueada")
````
